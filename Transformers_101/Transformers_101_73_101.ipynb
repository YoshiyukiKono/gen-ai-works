# 85本目

```python
from transformers import MobileBertTokenizer, MobileBertForMaskedLM
import torch

tokenizer = MobileBertTokenizer.from_pretrained('google/mobilebert-uncased')
model = MobileBertForMaskedLM.from_pretrained('google/mobilebert-uncased')

inputs = tokenizer("The capital of Japan is [MASK].", return_tensors="pt")
labels = tokenizer("The capital of Japan is Tokyo.", return_tensors="pt")["input_ids"]

outputs = model(**inputs, labels=labels)
loss = outputs.loss
logits = outputs.logits
print(' '.join([tokenizer.decode(i.item()).replace(" ", "") for i in logits.argmax(-1)[0]][1:-1]))
```

```python
from transformers import BertTokenizer, BertForMaskedLM
import torch

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertForMaskedLM.from_pretrained('bert-base-uncased')

inputs = tokenizer("The capital of Japan is [MASK].", return_tensors="pt")
labels = tokenizer("The capital of Japan is Tokyo.", return_tensors="pt")["input_ids"]

outputs = model(**inputs, labels=labels)
loss = outputs.loss
logits = outputs.logits
print(' '.join([tokenizer.decode(i.item()).replace(" ", "") for i in logits.argmax(-1)[0]][1:-1]))
```

```python
def get_cls_vector(sample_text):
  hidden_state = get_feature(sample_text, padding=True, truncation=True, max_length=512)
  cls_vec = np.array(hidden_state)[0, 0]
  return cls_vec

ml = moderna_text.split("\n")
pfzl = pfizer_text.split("\n")
ml = [text for text in ml if text != '']
pfzl = [text for text in pfzl if text != '']
```

# 100本目

```python
pfizer_dics = {}
for text in pfzl:
  for token in tokenizer.tokenize(text):
    if len(token) > 1:
      if token not in pfizer_dics:
        pfizer_dics[token] = 1
      else:
        pfizer_dics[token] += 1
sorted(pfizer_dics.items(), key=lambda x:-x[1])[:30]
```

```python
ml_dics = {}
for text in ml:
  for token in tokenizer.tokenize(text):
    if len(token) > 1:
      if token not in ml_dics:
        ml_dics[token] = 1
      else:
        ml_dics[token] += 1
sorted(ml_dics.items(), key=lambda x:-x[1])[:30]
```

# 101本目

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, random_state=0)
moderna_vecs_reduced = tsne.fit_transform(moderna_vecs)
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(13, 7))
plt.scatter(moderna_vecs_reduced[:, 0], moderna_vecs_reduced[:, 1],  c=[i for i in range(len(ml))])
plt.show()
```
